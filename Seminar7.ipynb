```python
from scipy.stats import f
import math
from scipy.stats import t

def corr_s(arr,n,alpha):
    res = np.array(arr)
    for i in range(np.shape(arr)[0]):
        for j in range(np.shape(arr)[0]):
            if i != j:
                tn = math.fabs(res[i,j]) * math.sqrt(n-2) / math.sqrt(1 - res[i,j]**2)
                tk = t.ppf(1 - alpha, n-2)
                if tn < tk:
                    res[i,j] = 0
    res = pd.DataFrame(res)
    res.index = arr.index;res.columns = arr.columns
    return res

C_s = corr_s(corr,13,0.05).style.background_gradient(cmap='viridis')
display(C_s)
```

```python
def Txy(x):
    from collections import Counter
    count = list(Counter(x).values())
    res = 0
    for i in count:
        res += i**3 - i
    return 1/2*res

n = len(dfr.index)
Tx = Txy(dfr['sl'])
Ty = Txy(dfr['sw'])
Pxy = (1/6*(n**3-n)-((dfr['sl']-dfr['sw'])**2).sum()-Tx-Ty)/(math.sqrt((1/6*(n**3-n)-2*Tx)*(1/6*(n**3-n)-2*Ty)))
Pxy
```

```python
def MyKendall(dataFrame,columns):
    P=0
    Q=0

    df_rank = dataFrame[columns].rank()
    df_rank_sort_sl = df_rank.sort_values(by=columns[0])
    
    for i in range(df_rank_sort_sl.shape[0]):
        df_1 = df_rank_sort_sl.iloc[i+1:,1]
        P+= df_1[df_1 > df_rank_sort_sl.iloc[i,1]].count()
    
    for i in range(df_rank_sort_sl.shape[0]):
        df_1 = df_rank_sort_sl.iloc[i+1:,1]
        Q+= df_1[df_1 < df_rank_sort_sl.iloc[i,1]].count()
    
    S=P-Q
    n = df_rank_sort_sl.shape[0]
    
    return 2*S/(n*(n-1))

print(MyKendall(MyData[['sl','sw']],['sl','sw']))
```

```python
def corrKendall(dataFrame):
    res = np.zeros((dataFrame.shape[1],dataFrame.shape[1]))
    for i in range(dataFrame.shape[1]):
        for j in range(dataFrame.shape[1]):
            if(i == j):
                res[i,j] = 1
            else:
                res[i,j] = MyKendall(dataFrame, [dataFrame.columns[i],dataFrame.columns[j]])
    res = pd.DataFrame(res)
    res.index = dataFrame.columns
    res.columns = dataFrame.columns
    return res
```

```python
def koefAss(frame):
    kau = (frame.iloc[0, 0] * frame.iloc[1, 1] - frame.iloc[1, 0] * frame.iloc[0, 1]) / (frame.iloc[0, 0] * frame.iloc[1, 1] + frame.iloc[1, 0] * frame.iloc[0, 1])
    return kau

def koefkont(frame):
    chisl = frame.iloc[0, 0] * frame.iloc[1, 1] - frame.iloc[1, 0] * frame.iloc[0, 1]
    znam = math.sqrt((frame.iloc[0, 0] + frame.iloc[0, 1])*(frame.iloc[0, 1] + frame.iloc[1, 1])*(frame.iloc[0, 0] + frame.iloc[1, 0])*(frame.iloc[1, 0] + frame.iloc[1, 1]))
    res = chisl/znam
    return res
```
