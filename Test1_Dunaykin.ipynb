```python
#Задача №1
n = int(input('Введите число:'))
l = [1, 2, 3, 4, 5, 11, 53, 66]
def func1(l, n):
    temp = list()
    for i in l:
        if i % n == 0:
            temp.append(i)
    return temp
print('Выходной список:', func1(l, n))
```

```python
#Задача №2
l1 = list(range(1,11))
print(l1)
def func2(l1):
    ll1 = list()
    for i in range(len(l1)):
        if i % 3 == 0:
            continue
        elif i % 5 == 0:
            ll1.append(l1[i] * 2)
        else:
            ll1.append(l1[i])
    return ll1
print('Итог:', func2(l1))
```

```python
#Задача №3
z = int(input('Введите число:'))
l2 = [1, 2, 455, 6, 6, 1, 6]
def func3(l2, z):
    return l2.count(z)
print('Кол-во повторений:', func3(l2, z))
```

```python
#Задача №4
x = list(range(1, 13))
print('Старт:', x)
def func4(x):
    i = 0
    while (i + 2) <= len(x):
        x[i], x[i + 2] = x[i + 2], x[i]
        i += 3
    return x
print('Вывод:', func4(x))
```

```python
#Задача №5
s1 = [1, 2, 5, 6, 7, 3]
s2 = [6, 7, 3, 7, 1, -6]
def func5(s1, s2):
    s3 = list()
    i = 0
    while i < len(s1 + s2)/2:
        s3.append(s1[i] - s2[i])
        i += 1
    return s3
print('Выходные данные:', func5(s1, s2))
```

```python
#Задача №6
L = [1, 2, 5, 6, 4, 3, 3, 5, 4, 5, 7]
def func6(L):
    ss = list()
    for i in L:
        ss.append(L.count(i))
    return ss
print('Вывод:', func6(L))
```

```python
#Задача №7
LL = [-10, 16, 13, 3, 0, -16, 1, -6, 0, 4, 0]
def func7(LL):
    new_L = list()
    for i in LL:
        if i > 0:
            new_L.append(i)
    new_L.sort()
    return new_L
print('Вывод:', func7(LL))
```

```python
#Задача №8
N = int(input('Введите число:'))
Q = list(range(0, N - 2))
print('Исходный список:', Q)
def func8(N, Q):
    temp1 = list()
    for i in range(0, N):
        if i not in Q:
            temp1.append(i)
    return temp1
print('Результат:', func8(N, Q))
```
