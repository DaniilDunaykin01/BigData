```python
#Задача №1
n = int(input('Введите число:'))
l = [1, 2, 3, 4, 5, 11, 53, 66]
def func1(l, n):
    temp = list()
    for i in l:
        if i % n == 0:
            temp.append(i)
    return temp
print('Выходной список:', func1(l, n))
```

```python
#Задача №2
l1 = list(range(1,11))
print(l1)
def func2(l1):
    ll1 = list()
    for i in range(len(l1)):
        if (i + 1) %5 == 0:
            ll1.append(l1[i] * 2)
        elif (i + 1) %3 == 0:
            continue  
        elif (i + 1) % 5 != 0 or (i + 1) % 3 != 0:
            ll1.append(l1[i])
    return ll1
print('Итог:', func2(l1))
```

```python
#Задача №4
x = list(range(1, 13))
print('Старт:', x)
def func4(x):
    i = 0
    while (i + 2) <= len(x):
        x[i], x[i + 2] = x[i + 2], x[i]
        i += 3
    return x
print('Вывод:', func4(x))
```

```python
#Задача №5
s1 = [11, 2, 3, 47, 55, 77]
s2 = [55, 76, 1, -67, 99, 22]
def func5(s1, s2):
    s3 = list()
    i = 0
    while i < len(s1 + s2)/2:
        s3.append(s1[i] - s2[i])
        i += 1
    return s3
print('Выходные данные:', func5(s1, s2))
```

```python
#Задача №6
L = [1, 2, 5, 6, 4, 3, 3, 5, 4, 5, 7]
def func6(L):
    ss = list()
    works = list()
    for i in L:
        if i in works:
            continue
        ss.append(L.count(i))
        works.append(i)
    return ss
print('Вывод:', func6(L))
```

```python
#Задача №7
LL = [-10, 16, 13, 3, 0, -16, 1, -6, 0, 4, 0]
def func7(LL):
    new_L = list()
    for i in LL:
        if i > 0:
            new_L.append(i)
    new_L.sort()
    return new_L
print('Вывод:', func7(LL))
```

```python
#Задача №8
N = int(input('Введите число:'))
Q = list(range(0, N - 3))
print('Исходный список:', Q)
def func8(N, Q):
    temp1 = list()
    for i in range(0, N):
        if i not in Q:
            temp1.append(i)
    return temp1
print('Результат:', func8(N, Q))
```

```python
#Из папки на github с текущим заданием загрузить файл data_1.csv
import numpy as np
import pandas as pd
import matplotlib

data=pd.read_csv("https://raw.githubusercontent.com/junaart/ForStudents/master/Big_Data/Seminar_4/data_1.csv")
display(data)
```

```python
#Cформировать два отдельных DataFrame из мужчин и женщин. Для женщин рассчитать и добавить в соответствующий DataFrame
#столбец возраста. Для мужчин рассчитать и добавить в соответствующий DataFrame столбец стажа.
women_frame = pd.DataFrame(columns = data.columns)
women_frame = data[data['Пол'] == 'ж'].copy()
display(women_frame)
men_frame = pd.DataFrame(columns = data.columns)
men_frame = data[data['Пол'] == 'м'].copy()
display(men_frame)

for label in women_frame.index:
    women_frame.loc[label,'Возраст'] = 2021 - int(women_frame.loc[label,'Дата рождения'][6:10])
display(women_frame)

for label in men_frame.index:
    men_frame.loc[label,'Стаж'] = 2021 - int(men_frame.loc[label,'Дата рождения'][6:10])
display(men_frame)
```

```python
#Отобрать только сотрудников, имеющих детей и оклад, меньше среднего оклада сотрудников.
temp = data.Оклад.mean()
print(temp)
display(data[(data['Оклад'] < temp) & (data['Кол-во детей'] > 0)])
```

```python
#Выяснить, есть ли в списке юбиляры (кому в текущем году исполнилось количество лет, оканчивающееся на нуль или пять),
#получить этот список в виде DataFrame.
def function1(argument):
    temp = list()
    for i in argument:
        temp.append(int(i[6:10]))
    return pd.Series(temp)

display(data[(2021 - function1(data['Дата рождения'])) % 5 == 0])
```

```python
#Получить DataFrame, в котором рассчитывается количество сотрудников по категориям:
#женат, замужем, холост, незамужем, разведен, разведена. Полученный DataFrame сохранить на github
#в формате csv с указанием ссылки.
new_df = pd.DataFrame(columns = ['Женат','Замужем','Холост','Незамужем','Разведен','Разведена'])
new_df.loc[0, 'Женат'] = data[data['Семейное положение'] == 'женат']['Семейное положение'].count()
new_df.loc[0, 'Замужем'] = data[data['Семейное положение'] == 'замужем']['Семейное положение'].count()
new_df.loc[0, 'Холост'] = data[data['Семейное положение'] == 'холост']['Семейное положение'].count()
new_df.loc[0, 'Незамужем'] = data[data['Семейное положение'] == 'незамужем']['Семейное положение'].count()
new_df.loc[0, 'Разведен'] = data[data['Семейное положение'] == 'разведен']['Семейное положение'].count()
new_df.loc[0, 'Разведена'] = data[data['Семейное положение'] == 'разведена']['Семейное положение'].count()
display(new_df)
```
